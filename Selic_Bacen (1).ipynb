# Coleta de dados da Selic

Obtenção da Taxa Selic Meta utilizando Python, por meio da biblioteca {python-bcb} em conjunto com o módulo sgs, com objetivo de importar os dados do Sistema Gerenciador de Séries Temporais do Banco Central (SGS).

Estudo inspirado pelo Boletim da Análise Macro 

## Imports

```python
from bcb import sgs
import pandas as pd
from matplotlib import pyplot as plt
```

## Coleta de dados

No sites do SGS (https://www3.bcb.gov.br/sgspub/localizarseries/localizarSeries.do?method=prepararTelaLocalizarSeries) o código 432 indica a série da Taxa Meta Selic, do período 05/03/1999 até 15/06/2022. Por meio da função sgs.get() é possível colher os dados na série histórica. Entretanto, vou coletar os dados a parti de 2015, antes do impedimento da ex-presidente Dilma.

```python
selic = sgs.get(('selic', 432), start = '2015-01-01')

selic.plot(figsize = (10,8))
plt.title("Taxa de juros - Meta Selic definida pelo Copom - %a.a")
plt.style.use('fast')
plt.show()
```

## Conclusão

Durante o governo Dilma, foram observados as maiores taxas de juros. Após o impeachment, houve um melhor controle da taxa básica de juros. Tal instrumento é importante para fazer a política monetária no Brasil (e em outros países). Após o governo Temer até 2021, foram conhecidas as taxas mais baixas desde o início do plano real. Mesmo com a pandemia as taxas se mantiveram, mas após o primeiro trimestre de 2021, as taxas voltaram a elevar-se dado a expansão monetária (auxílio Brasil, por exemplo) que ocorreu não somente no país, mas também no mundo todo. A selic é importante para o cidadão "comum", dado que afata financiamentos, emprestimo e até investimentos em renda fixa.
